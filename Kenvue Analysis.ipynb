```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('~/Desktop/kenvuedata.csv')
```

```python
# Converting the 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# here we create a new column for month
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

# identifying our specific 'Need States' value #5
need_states_value = 'Need State 5'
filtered_df = df[df['Need States'] == 'Need State 5']

# Grouping by month to calculate average total shares
avg_sales_by_month = filtered_df.groupby('month')['Total Sales'].mean()

# visualize the results
print(avg_sales_by_month)
```

```python
# plotting the data
plt.scatter(avg_sales_by_month.index, avg_sales_by_month.values)
plt.xlabel('Month')
plt.ylabel('Average Total Sales')
plt.title(f'Average Total Sales per Month for {need_states_value}')
plt.show()
```

```python
# When we just take a look at overall averages for each month for all years, the output data is very skewed.
# For further insight we will do a year-by-year breakdown of monthly averages.


df['month'] = df['date'].dt.month
```
